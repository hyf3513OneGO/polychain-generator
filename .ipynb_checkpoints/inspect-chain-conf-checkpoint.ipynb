```python
from rdkit import Chem
from rdkit.Chem import AllChem
import py3Dmol
import os

# 读取SDF文件
result_folder = "results/dev_test-1/task_4_repeat-5_25000_2.0_298.0k_gaff-2.11"
sdf_file = os.path.join(result_folder, "init_chain.sdf")
smiles = ""
smiles_path = os.path.join(result_folder,"psmiles.txt")
with open(smiles_path,"r") as f:
    smiles = f.readline()
mol = Chem.MolFromMolFile(sdf_file, removeHs=False)  # 保留H

# 去除所有氢原子
mol_noH = Chem.RemoveHs(mol)

# 生成 MolBlock（包含3D坐标）
block = Chem.MolToMolBlock(mol_noH)

# 获取构象
conf = mol_noH.GetConformer()

# 初始化 py3Dmol 可视化
view = py3Dmol.view(width=500, height=500)
view.addModel(block, "mol")
view.setStyle({'stick': {}})
view.zoomTo()

# 添加原子编号标签
for atom in mol_noH.GetAtoms():
    idx = atom.GetIdx()
    pos = conf.GetAtomPosition(idx)
    view.addLabel(str(idx), {
        'position': {'x': pos.x, 'y': pos.y, 'z': pos.z},
        'backgroundColor': 'none',
        'fontColor': 'white',
        'fontSize': 16,
        'showBackground': True
    })

view.show()
print(smiles)
```

```python
monomer = Chem.MolFromSmiles(smiles)
from rdkit.Chem import Draw
monomer_noH = Chem.RemoveHs(monomer)
for atom in monomer_noH.GetAtoms():
    idx = atom.GetIdx()
    symbol = atom.GetSymbol()
    print(idx,symbol)
Draw.MolToImage(monomer, size=(300, 300))
```
